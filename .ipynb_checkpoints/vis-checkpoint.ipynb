```python
from data.DirectBags import DirectBags
import  torch
from torch.utils.data import DataLoader, Dataset
import torch.nn as nn
import os
from torchvision.transforms import transforms
from data.DirectBags import DirectBags
from models.graph_attention import H_Attention_Graph
from tqdm import tqdm
from matplotlib import pyplot as plt
```

```python
def inference():

    model = H_Attention_Graph(t=0.5).cuda()
    model_path = '/media/hhy/data/gcn_results/myALNM10_newauc0/ckp/net.ckpt41.pth'
    model.load_state_dict(torch.load(model_path)['net'])
    ## load
    ## change to BM

    test_transform = transforms.Compose([
        transforms.Resize((112,112)),
        transforms.ToTensor()
    ])

    root = '/media/hhy/data/USdata/MergePhase1/vis'

    bag = DirectBags(root, pre_transform=test_transform,sub_list=[0])
    loader = DataLoader(bag, batch_size=1, shuffle= False, num_workers = 8)
    model.eval()
    x_list = []
    weights_list = []
    num_list = []
    test_list =[]
    tf_list =[]
    path_list = []
    for idx, (data,img_info, label, idx_list, path) in enumerate(tqdm(loader,ascii=True,ncols=60)):
        data = data.cuda()
        label = label.cuda()
        img_info = img_info.cuda()
        prob_label, predicted_label,_, weights = model(data,img_info,label,idx_list)
        #print(predicted_label)
        #print(label)
        #print(path)
        #print(weights)
        weights = weights.squeeze(0).tolist()
        if len(weights) == 1:
            continue
        x = len(weights) * [idx]
        #print(x)
        #print(weights)
        x_list.extend(x)
        weights_list.extend(weights)
        num_list.append(len(weights))
        test_list.append(weights)
        path_list.append(path)
        #plt.scatter(x,weights)
        #print(label.tolist()[0])
        #print(predicted_label.tolist()[0][0])
        
        tf_list.append(label.tolist()[0] == predicted_label.tolist()[0][0])
    
    return x_list, weights_list,num_list, test_list,tf_list, path_list

```

```python
x, w,n,t,tf,path_list = inference()

```
